```python
import numpy as np
from time import sleep
import matplotlib.pyplot as plt
%matplotlib inline
```

### SIR

```python
i_rate = 1.0
r_rate = 0.2
```

```python
SIR = np.array([100000,1,0])

N = 100001
```

```python
trans = np.zeros((3,3))
def iterate(i_rate, r_rate, SIR, N):
    trans[0,1] = (-i_rate)*SIR[0]/N
    trans[1,1] = ((i_rate)*SIR[0]/N) - r_rate
    trans[2,1] = r_rate
    #print(trans)
    return SIR + np.matmul(trans, SIR)
    
```

```python
print(SIR)
epidemic = [SIR]
for i in range(100):
    SIR = iterate(i_rate, r_rate, SIR, N)
    epidemic.append(SIR)
    print(SIR)
```

```python
epidemic = np.array(epidemic)
```

```python
plt.plot(epidemic)
```

### SIRD

```python
i_rate = 1.0
r_rate = 0.2
d_rate = 0.01
```

```python
SIRD = np.array([10000,1,0,0])
```

```python
N = 10001
```

```python
trans = np.zeros((4,4))
def iterate(i_rate, r_rate, d_rate, SIRD, N):
    trans[0,1] = (-i_rate)*SIRD[0]/N
    trans[1,1] = ((i_rate)*SIRD[0]/N) - r_rate - d_rate
    trans[2,1] = r_rate
    trans[3,1] = d_rate
    #print(trans)
    return SIRD + np.matmul(trans, SIRD)
    
```

```python
print(SIRD)
epidemic = [SIRD]
for i in range(100):
    SIRD = iterate(i_rate, r_rate, d_rate, SIRD, N)
    epidemic.append(SIRD)
    print(SIRD)
```

```python
epidemic = np.array(epidemic)
```

```python
plt.plot(epidemic)
```
